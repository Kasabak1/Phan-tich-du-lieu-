### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('chipotle.tsv', sep='\t')
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

### Step 4. How many products cost more than $10.00?

```python
chipo_filtered = chipo.drop_duplicates(['item_name', 'choice_description', 'item_price'])
products_above_10 = chipo_filtered[chipo_filtered['item_price'] > 10.00]
print("Số sản phẩm có giá lớn hơn $10.00 là:", len(products_above_10))
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
item_prices = chipo[['item_name', 'item_price']].drop_duplicates()
print(item_prices)
```

### Step 6. Sort by the name of the item

```python
chipo_sorted = chipo.sort_values(by='item_name')
print(chipo_sorted.head())
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive = chipo.sort_values(by='item_price', ascending=False)
quantity_of_most_expensive = most_expensive['quantity'].iloc[0]
print("Số lượng của mặt hàng đắt nhất được đặt là:", quantity_of_most_expensive)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
print("Số lần Veggie Salad Bowl được đặt là:", len(veggie_salad_orders))
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_multiple = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
print("Số lần Canned Soda được đặt nhiều hơn 1 lon là:", len(canned_soda_multiple))
```
