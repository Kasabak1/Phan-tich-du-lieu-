### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Crime_Rates_1960_2014.csv).

### Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv('crime.csv', sep=',')
```

### Step 4. What is the type of the columns?

```python
print(crime.dtypes)
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime = crime.set_index('Year')
print(crime.head())
```

### Step 7. Delete the Total column

```python
crime = crime.drop('Total', axis=1, errors='ignore')
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crimes_decade = crime.resample('10YS').sum()
population_max = crime['Population'].resample('10YS').max()
crimes_decade['Population'] = population_max
print(crimes_decade)
```

### Step 9. What is the most dangerous decade to live in the US?

```python
crimes_decade['Total_Crimes'] = crimes_decade.drop('Population', axis=1).sum(axis=1)
most_dangerous = crimes_decade['Total_Crimes'].idxmax()
print("Thập kỷ nguy hiểm nhất bắt đầu từ năm:", most_dangerous.year)
```
