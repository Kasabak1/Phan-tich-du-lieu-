### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/pandas/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv('drinks.csv', )
drinks.head(5)
```

### Step 4. Which continent drinks more beer on average?

```python
beer_avg = drinks.groupby('continent')['beer_servings'].mean()
beer_avg_sorted = beer_avg.sort_values(ascending=False)
print("Châu lục uống nhiều bia nhất trên mức trung bình là:", beer_avg_sorted.index[0])
print("\nChi tiết mức tiêu thụ bia trung bình:\n", beer_avg_sorted)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine_stats = drinks.groupby('continent')['wine_servings'].describe()
print(wine_stats)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
mean_alcohol = drinks.groupby('continent').mean(numeric_only=True)
print(mean_alcohol)
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
median_alcohol = drinks.groupby('continent').median(numeric_only=True)
print(median_alcohol)
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])
print(spirit_stats)
print("\nKiểu dữ liệu của spirit_stats là:", type(spirit_stats))
```
