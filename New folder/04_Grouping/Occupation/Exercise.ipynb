### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('occupation.csv', sep='|', index_col='user_id')
print(users.head())
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occ = users.groupby('occupation')['age'].mean()
print(mean_age_per_occ)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_ratio = (users['gender'] == 'M').groupby(users['occupation']).mean()
male_ratio_sorted = male_ratio.sort_values(ascending=False)
print(male_ratio_sorted)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_age)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_combo = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_combo)   
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_counts = users.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
total_per_occ = gender_counts.sum(axis=1)
gender_percentages = gender_counts.div(total_per_occ, axis=0) * 100
gender_percentages = round(gender_percentages, 2)
print(gender_percentages)
```
