### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('chipotle.tsv', sep='\t')
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

### Step 4. See the first 10 entries

```python
print(chipo.head(10))
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_5_items = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)

top_5_items.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Top 5 món ăn được đặt nhiều nhất tại Chipotle')
plt.xlabel('Tên món ăn')
plt.ylabel('Tổng số lượng đã đặt')

plt.xticks(rotation=45, ha='right') 

plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
orders = chipo.groupby('order_id').sum(numeric_only=True)
plt.scatter(x=orders['item_price'], y=orders['quantity'], alpha=0.5, color='green')
plt.title('Số lượng món ăn theo Giá trị đơn hàng')
plt.xlabel('Giá trị đơn hàng ($)')
plt.ylabel('Số lượng món ăn')
plt.show()
```
