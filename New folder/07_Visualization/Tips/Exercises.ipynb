### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('tips.csv', sep=',')
print(tips.head())
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop('Unnamed: 0', axis=1)

print(tips.head())
```

### Step 5. Plot the total_bill column histogram

```python
# Vẽ histogram cho cột hóa đơn tổng (total_bill)
sns.histplot(tips['total_bill'], kde=False, color='skyblue', bins=20)

plt.title('Phân phối Giá trị Hóa đơn')
plt.xlabel('Tổng Hóa đơn ($)')
plt.ylabel('Tần suất')

plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, color='coral')

plt.title('Mối quan hệ giữa Hóa đơn và Tiền bo')
plt.xlabel('Tổng Hóa đơn ($)')
plt.ylabel('Tiền bo ($)')

plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips)

plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)

plt.title('Tổng hóa đơn theo Ngày trong tuần')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(x='tip', y='day', hue='sex', data=tips, jitter=True, palette='Set2')

plt.title('Tiền bo theo Ngày, phân biệt theo Giới tính')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips, palette='pastel')

plt.title('Phân phối Hóa đơn theo Ngày và Bữa ăn')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
g = sns.FacetGrid(tips, col='time', height=5)
g.map(sns.histplot, 'tip', bins=15, color='teal')

plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
sns.relplot(
    x='total_bill', 
    y='tip', 
    col='sex', 
    hue='smoker', 
    kind='scatter', 
    data=tips, 
    palette='muted'
)

plt.show()
```
