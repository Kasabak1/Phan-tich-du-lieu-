# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('titanic.csv')
print(titanic.head())
```

### Step 4. Set PassengerId as the index 

```python
titanic = titanic.set_index('PassengerId')
print(titanic.head())
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
gender_counts = titanic['Sex'].value_counts()

plt.pie(
    gender_counts, 
    labels=['Nam (Male)', 'Nữ (Female)'], 
    autopct='%1.1f%%', 
    colors=['lightblue', 'lightpink'], 
    startangle=90, 
    explode=(0.05, 0) # Tách nhẹ 2 phần ra cho đẹp
)

plt.title('Tỷ lệ hành khách Nam / Nữ trên tàu Titanic')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.scatterplot(
    x='Age', 
    y='Fare', 
    hue='Sex', 
    data=titanic, 
    palette={'male': 'royalblue', 'female': 'crimson'}, 
    alpha=0.6
)

plt.title('Mối quan hệ giữa Giá vé và Tuổi theo Giới tính')
plt.xlabel('Tuổi (Age)')
plt.ylabel('Giá vé (Fare)')

plt.show()
```

### Step 7. How many people survived?

```python
survived_count = titanic['Survived'].sum()
print(f"Số người sống sót sau thảm họa là: {survived_count} người")
```

### Step 8. Create a histogram with the Fare payed

```python
sns.histplot(titanic['Fare'], bins=50, color='seagreen')

plt.title('Phân phối Giá vé (Fare) trên tàu Titanic')
plt.xlabel('Giá vé ($)')
plt.ylabel('Số lượng hành khách')

plt.show()
```
