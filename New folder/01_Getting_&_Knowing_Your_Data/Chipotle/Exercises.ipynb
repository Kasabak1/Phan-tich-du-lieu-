### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv(url, sep='\t')
```

### Step 4. See the first 10 entries

```python
print(chipo.head(10))
```

### Step 5. What is the number of observations in the dataset?

```python
print(len(chipo))
```

### Step 6. What is the number of columns in the dataset?

```python
print(chipo.shape[1])
```

### Step 7. Print the name of all the columns.

```python
print(chipo.columns.tolist())
```

### Step 8. How is the dataset indexed?

```python
print(chipo.index)
```

### Step 9. Which was the most-ordered item? 

```python
c = chipo.groupby('item_name').sum()
c = c.sort_values(['quantity'], ascending=False) 
most_ordered = c.head(1)   
print(f"\n---Mục được đặt nhiều nhất là '{most_ordered.index[0]}' ---")
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
print(f"\n---Mục được đặt nhiều nhất là '{most_ordered.index[0]}' với {most_ordered['quantity'].iloc[0]} phần ---")
```

### Step 11. What was the most ordered item in the choice_description column?

```python
c_choice = chipo.groupby('choice_description').sum()
c_choice = c_choice.sort_values(['quantity'], ascending=False)
print(c_choice.head(1))
```

### Step 12. How many items were orderd in total?

```python
total_items_ordered = chipo['quantity'].sum()
print(f"\n--- Tổng số mặt hàng đã đặt: {total_items_ordered} ---")
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print(chipo['item_price'].dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
dollarizer = lambda x: float(x[1:-1])
chipo['item_price'] = chipo['item_price'].apply(dollarizer)
```

#### Step 13.c. Check the item price type

```python
print(chipo['item_price'].dtype)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print(f"\n--- Tổng doanh thu: ${revenue:.2f} ---")
```

### Step 15. How many orders were made in the period?

```python
orders = chipo['order_id'].value_counts().count()
print(f"\n--- Số đơn hàng đã thực hiện: {orders} ---")   
```

### Step 16. What is the average revenue amount per order?

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
order_grouped = chipo.groupby('order_id').sum()
avg_revenue = order_grouped['revenue'].mean()
print(f"\n--- Doanh thu trung bình mỗi đơn hàng: ${avg_revenue:.2f} ---")
```
